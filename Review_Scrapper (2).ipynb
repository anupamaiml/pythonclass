```python
from bs4 import BeautifulSoup as bs
import requests
#pip install BeautifulSoup
#pip install requests
```

```python
Url = 'https://www.flipkart.com/samsung-galaxy-s9-midnight-black-64-gb/p/itmf33a69rpszgzn?pid=MOBF2VWVBGCT5QQN&lid=LSTMOBF2VWVBGCT5QQN0ZJFUP&marketplace=FLIPKART&srno=s_1_1&otracker=search&otracker1=search&fm=SEARCH&iid=b99d2b09-48b3-4c1d-a31d-b20ed351d6f1.MOBF2VWVBGCT5QQN.SEARCH&ppt=sp&ppn=sp&ssid=aed6k7dn740000001592683142789&qH=fe546279a62683de'
```

```python
req_data = requests.get(Url)
```

```python
req_data
```

```python
review_soup = bs(req_data.content)
```

```python
review_soup
```

```python
all_reviews = review_soup.find_all('div', {'class': ['col _39LH-M']})
```

```python
all_reviews
```

```python
rating_list = []
header_list = []
content_list = []
user_list = []
likes_dislikes_list = []

for r in all_reviews:
    rating = r.find_all('div', {'class': 'hGSR34 E_uFuv'})
    text = [e.get_text() for e in rating]
    rating_list.append(text)
    header = r.find_all('p', {'class': '_2xg6Ul'})
    h_text = [e.get_text() for e in header]
    header_list.append(h_text)
    content = r.find_all('div', {'class': 'qwjRop'})
    c_text = [e.get_text() for e in content]
    content_list.append(c_text)
    user = r.find_all('p', {'class': '_3LYOAd _3sxSiS'})
    u_text = [e.get_text() for e in user]
    user_list.append(u_text)
    l_and_d = r.find_all('span', {'class': '_1_BQL8'})
    ld_text = [e.get_text() for e in l_and_d]
    likes_dislikes_list.append(ld_text)
```
